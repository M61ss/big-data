## Pivot Table

```python
dataset.pivot_table(index=['InternetService', 'Contract'], values=['Churn'], aggfunc='count')
```

### Count-to-Percentage

Global

```python
df = dataset.copy()
df['Churn'] = df['Churn'].map({'Yes' : 1, 'No' : 0})
pivot = df.pivot_table(index=['InternetService'], columns=['Contract'], values=['Churn'], aggfunc='mean', observed=True)
pivot
```

Local

```python
pivot = dataset.pivot_table(index=['InternetService', 'Contract'], values=['Churn'], aggfunc='count')
pivot['Churn'] / pivot.groupby('InternetService')['Churn'].sum()
```

## Plots

### Hist

Useful to observe the distribution of a single feature

```python
df['MonthlyCharges'].hist(bins=30)
```

### Bar

Useful to observe values or distribution of a feature w.r.t. another one

```python
df = dataset.copy()
df['Churn'] = df['Churn'].map({'Yes' : 1, 'No' : 0})
df.groupby('InternetService')['Churn'].mean().plot.bar()    # this visualize the mean churn wrt internet service, not a distribution!
```

It is also possible to observe the distribution of every value of a feature w.r.t another one

```python
df = dataset.copy()
df.groupby(['InternetService', 'Churn']).size().unstack(fill_value=0).plot.bar()
```

```python
# Equivalent to the previous, but easier to understand
pd.crosstab(df['InternetService'], df['Contract']).plot.bar()
```
